# Сортировка Counting sort для целых чисел

```python
def counting_sort(arr):
    max_val = max(arr)
    min_val = min(arr)
    range_of_elements = max_val - min_val + 1
    count_arr = [0] * range_of_elements
    output_arr = [0] * len(arr)

    
    for i in range(len(arr)):
        count_arr[arr[i] - min_val] += 1

    
    for i in range(1, len(count_arr)):
        count_arr[i] += count_arr[i - 1]

    
    for i in range(len(arr) - 1, -1, -1):
        output_arr[count_arr[arr[i] - min_val] - 1] = arr[i]
        count_arr[arr[i] - min_val] -= 1

   
    for i in range(len(arr)):
        arr[i] = output_arr[i]

    return arr



print("Исходный массив:", random_Cel)
print("Отсортированный массив:", counting_sort(random_Cel))
```

# Сортировка merge sort Для случайных чисел из диапазона -1:1

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2  # Находим середину массива
        L = arr[:mid]        # Делим массив на две части
        R = arr[mid:]

        merge_sort(L)  
        merge_sort(R)  

        i = j = k = 0

        
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

    return arr

print("Исходный массив:", random_natur)
print("Отсортированный массив:", merge_sort(random_natur))
```

# Сортировка Комплексных чисел Пузырьком

```python
def bubble_sort_complex(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if abs(arr[j]) > abs(arr[j + 1]):
                arr[j], arr[j + 1] = arr[j + 1], arr[j]




bubble_sort_complex(complex_points)
print("Отсортированный массив комплексных чисел:")
for num in complex_points:
    print(num)
```

# Сортировка Текста

```python
def counting_sort(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 256  # Extended ASCII

    for i in range(n):
        index = ord(arr[i][exp]) if len(arr[i]) > exp else 0
        count[index] += 1

    for i in range(1, 256):
        count[i] += count[i - 1]

    i = n - 1
    while i >= 0:
        index = ord(arr[i][exp]) if len(arr[i]) > exp else 0
        output[count[index] - 1] = arr[i]
        count[index] -= 1
        i -= 1

    for i in range(n):
        arr[i] = output[i]


def radix_sort_words(arr):
    max_len = max(len(word) for word in arr)
    for exp in range(max_len - 1, -1, -1):
        counting_sort(arr, exp)



radix_sort_words(words)
print("Отсортированные слова:")
print(words)
```
