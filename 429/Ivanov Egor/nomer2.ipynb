# 2  Пузырек

```python
def bubble_sort(mass):
    n=len(mass)
    for i in range(n-1):
        flag=True
        for j in range(n-i-1):
            if mass[j]>mass[j+1]:
                flag=False
                mass[j], mass[j+1]=mass[j+1], mass[j]
        if flag:
            break
    return mass
```

```python
print(bubble_sort(z_numbers))
```

# 3 Рассческа

```python
def rascheska_sort(mass):
    n = len(mass)
    step=n
    while step > 1 or flag:
        flag=False
        i=0
        if step>1:
            step=int(step-step*0.2)
        while i+step<n:
            if mass[i]>mass[i+step]:
                flag=True
                mass[i], mass[i+step]= mass[i+step], mass[i]
            i+=step
    
    return mass
```

```python
print(rascheska_sort(n_numbers))
```

# 7 Гномья сортировка

```python
def gnom_sort(mass):
    n=len(mass)
    i, j =1, 2
    while i<n:
        if abs(mass[i-1])<abs(mass[i]):
            i=j
            j+=1
        else:
            mass[i-1], mass[i] = mass[i], mass[i-1]
            i-=1
            if i==0:
                i=j
                j+=1
    return mass
```

```python
print(gnom_sort(c_numbers))
```

# 11  Merge sort Сортировка слиянием

```python
def merge(mass1, mass2):
    n1, n2 = len(mass1), len(mass2)
    ans=[]
    i1, i2=0, 0
    for o in range(n1+n2):
        if i1==n1:
            ans+=[mass2[i2]]
            i2+=1
        elif i2==n2:
            ans+=[mass1[i1]]
            i1+=1
        elif mass1[i1]<mass2[i2]:
            ans+=[mass1[i1]]
            i1+=1
        else:
            ans+=[mass2[i2]]
            i2+=1

    return ans

def merge_sort(mass):
    n=len(mass)
    if n==1:
        return mass
    else:
        return merge(merge_sort(mass[:n//2]), merge_sort(mass[n//2+n%2:]))
```

```python
print(merge_sort(words))
```
