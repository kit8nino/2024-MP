# Задание 7
FIFO очередь, в которую можно добавлять строковые названия стройматериалов, вводимые с клавиатуры (до команды остановки), после введения - вывести все;

```python
from queue import Queue
q = Queue()
 
while True:
    temp=input()
    if temp:
        q.put(temp)
    else: break

while not q.empty():
    print(q.get())
```
